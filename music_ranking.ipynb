```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

music_title = []
music_artist = []
music_link = []
music_img = []
music_type = []
music_releasedate = []

for time in range(45,52):
    
    url = f'https://streetvoice.com/music/charts/weekly/2023/{time}/all/'
    
    res = requests.get(url)

    soup = BeautifulSoup(res.text, 'html.parser') 

    titles = soup.select('h4 a')
    artists = soup.select('h5 a')
    links = soup.find_all(class_='cover-block with-button img-lg mr-3') 
    imgs = soup.find_all('img')

    
    for title in titles:
        music_title.append(title.get_text().strip('\n\n').strip('    \n\n'))


    for artist in artists:
        music_artist.append(artist.get_text())
    

    for img in imgs:
        last_img = img['src']
        if last_img != 'https://akstatic.streetvoice.com/asset/images/sv-cat.gif':
            music_img.append(last_img)


    for link in links:
        last_links = 'https://streetvoice.com' + link.find('a')['href']
        music_link.append(last_links)
        
        music_res = requests.get(last_links)
        music_soup = BeautifulSoup(music_res.text, 'html.parser')
        
        types = music_soup.select('h3 a')
        releasedates = music_soup.find_all(class_='text-gray-light mb-2') 
        
        if types:
            music_type.append(types[0].text)
        else:
            music_type.append("")

        if releasedates:
            music_releasedate.append(releasedates[0].text.split(' ')[1])
        else:
            music_releasedate.append("")
```

```python
df = pd.DataFrame(  
{
    'music':music_title,
    'artist':music_artist,
    'link':music_link,
    'img':music_img,
    'type':music_type,
    'released_date':music_releasedate
})
df
```

```python
df.to_excel('final.xlsx', index=True)
```
