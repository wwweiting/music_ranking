# 留言爬取

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

file_path = 'final.xlsx'
df = pd.read_excel(file_path)

music_comment = []

for index, row in df.iterrows():
    url = row['link']
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser') 

    plays = soup.find_all(class_='text-read text-break mb-3') 

    for play in plays:
        chinese_text = re.sub(r'[^\u4e00-\u9fa5]+', '', play.text)
        music_comment.append(chinese_text)

music_comment
```

# 製作留言文字雲

```python
commentsAll = ''.join(music_comment)
commentsAll
```

```python
import jieba #結吧
Sentence = jieba.cut_for_search(commentsAll)  #.cut_for_search( )表示「搜尋引擎模式」
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#想刪除的文字
stopwords = {}.fromkeys(['的', '聽', '了', '後', '這', '到', '也', '吧', '我', '好', '很', '你', '是', '在', '真的', '啊', '都', '有', '首歌', '又', '來',
                         '上', '啦', '讓'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence = jieba.cut_for_search(commentsAll)

hash = {}

for item in Sentence:
 
    if item in stopwords: #將不要的字詞排除在雲之外
        continue
    
    if item in hash:
        hash[item] += 1   # hash[item] = hash[item] + 1
    else:
        hash[item] = 1

 
# 文字雲樣式設定


wc = WordCloud(font_path="TW-Kai-98_1.ttf",  #設置字體
               background_color="white",  #背景顏色
               max_words = 1000,  #文字雲顯示最大詞數
               stopwords=stopwords,  #停用字詞
               )   
 
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
 
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(50, 50), dpi = 300)

wc.to_file('WordCloud.jpg')
```
